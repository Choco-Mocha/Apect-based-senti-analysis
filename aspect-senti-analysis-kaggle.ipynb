```python
import pandas as pd
from langdetect import detect, LangDetectException
from googletrans import Translator, LANGUAGES
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from gensim import corpora
from gensim.models.ldamodel import LdaModel

# Download NLTK data files (you only need to do this once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

# Initialize the translator and lemmatizer
translator = Translator()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to detect language and translate if not English
def translate_comment(comment):
    try:
        print(f"Processing comment: {comment}")
        # Detect the language
        lang = detect(comment)
        print(f"Detected language: {lang}")
        # If the language is not English and is recognized
        if lang != 'en' and lang in LANGUAGES:
            translated = translator.translate(comment, src=lang, dest='en').text
            print(f"Translated comment: {translated}")
            return translated
        else:
            print("Comment is already in English or unrecognized language.")
            return comment
    except LangDetectException as e:
        # Handle language detection errors
        print(f"Language detection error: {e}")
        return f"Language detection error: {e}"
    except Exception as e:
        # Handle any other errors that may occur
        print(f"Error occurred: {e}")
        return f"Error: {e}"

# Function to tokenize a comment
def tokenize_comment(comment):
    try:
        print(f"Tokenizing comment: {comment}")
        tokens = word_tokenize(comment)
        print(f"Tokens: {tokens}")
        return tokens
    except Exception as e:
        print(f"Tokenization error: {e}")
        return f"Tokenization error: {e}"

# Function to lemmatize tokens
def lemmatize_tokens(tokens):
    try:
        print(f"Lemmatizing tokens: {tokens}")
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
        print(f"Lemmatized tokens: {lemmatized_tokens}")
        return lemmatized_tokens
    except Exception as e:
        print(f"Lemmatization error: {e}")
        return f"Lemmatization error: {e}"

# Function to remove stopwords and non-alphabetic tokens
def clean_tokens(tokens):
    try:
        print(f"Cleaning tokens: {tokens}")
        cleaned_tokens = [token for token in tokens if token.isalpha() and token.lower() not in stop_words]
        print(f"Cleaned tokens: {cleaned_tokens}")
        return cleaned_tokens
    except Exception as e:
        print(f"Cleaning error: {e}")
        return f"Cleaning error: {e}"

# Apply the translation function to the comments column
print("Translating comments...")
df_sho['translated_comments'] = df_sho['comments'].apply(translate_comment)

# Apply the tokenization function to the translated comments column
print("Tokenizing translated comments...")
df_sho['tokenized_comments'] = df_sho['translated_comments'].apply(tokenize_comment)

# Apply the lemmatization function to the tokenized comments column
print("Lemmatizing tokenized comments...")
df_sho['lemmatized_comments'] = df_sho['tokenized_comments'].apply(lemmatize_tokens)

# Apply the cleaning function to the lemmatized comments column
print("Cleaning lemmatized comments...")
df_sho['cleaned_comments'] = df_sho['lemmatized_comments'].apply(clean_tokens)

# Prepare data for LDA
print("Preparing data for LDA...")
df_sho['lda_ready'] = df_sho['cleaned_comments']

# Create a dictionary and corpus for LDA
dictionary = corpora.Dictionary(df_sho['lda_ready'])
corpus = [dictionary.doc2bow(text) for text in df_sho['lda_ready']]

# Build LDA model
print("Building LDA model...")
lda_model = LdaModel(corpus, num_topics=2, id2word=dictionary, passes=20)

# Print topics
print("LDA Topics:")
for idx, topic in lda_model.print_topics(num_words=10):
    print(f"Topic {idx}: {topic}")

# Display the DataFrame
print("Final DataFrame:")
print(df_sho)
```

```python
import pandas as pd

# Define positive and negative words/emojis
positive_keywords = ['good', 'great', 'excellent', 'love', 'like', 'amazing', 'ü§§', 'üòä', 'üòÉ', 'üëç']
negative_keywords = ['bad', 'terrible', 'awful', 'hate', 'dislike', 'üò¢', 'üò†', 'üëé']

def calculate_score(comment):
    # Convert comment to lowercase to match keywords
    comment_lower = comment.lower()
    # Initialize score
    score = 3  # Neutral by default
    # Check for positive keywords
    if any(word in comment_lower for word in positive_keywords):
        score = 5  # Very good
    # Check for negative keywords
    elif any(word in comment_lower for word in negative_keywords):
        score = 1  # Very bad
    
    return score

# Apply the function to the 'comments' column
df_sho['overall_score'] = df_sho['comments'].apply(calculate_score)

# Display the DataFrame with the new column
print(df_sho)
```
